```python
def pinball(y,q,alpha):
    return (y-q)*alpha*(y>=q) + (q-y)*(1-alpha)*(y<q)

def pinball_score(true, pred):
    score = list()
    for qu in range(10,100,10):
        score.append(pinball(y=true,
            # q=pred[:,48*(int(qu*0.1)-1):48*(int(qu*0.1))],
            q=pred[:,int(qu*0.1)-1,:],
            alpha=qu/100).mean())
    return sum(score)/len(score)
```
